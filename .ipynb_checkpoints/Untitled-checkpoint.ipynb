### 1a. Find all players in the database who played at Vanderbilt University.

#### Explored collegeplaying, salaries, people Dfs

```python
collegeplaying_df = pd.read_sql("select * from collegeplaying;", con=engine)
collegeplaying_df.head(2)
```

```python
salaries_df = pd.read_sql("select * from salaries;", con=engine)
salaries_df.head(2)
```

```python
people_df = pd.read_sql("select * from people;", con=engine)
people_df.head(2)
```

```python
### print(collegeplaying_df.dtypes)
### print(salaries_df.dtypes)
### print(people_df.dtypes)
```

### Merged collegeplaying, salaries, people Dfs, kept necessary columns only.

```python
cp_salaries_people_df = reduce(lambda x,y: pd.merge(x,y, on='playerid', how='outer'),[collegeplaying_df, salaries_df, people_df])
cp_salaries_people_df
cp_salaries_people_df_filtered = cp_salaries_people_df[["playerid", "schoolid","namegiven","namelast","salary" ]]
cp_salaries_people_df_filtered
```

### Found all Vandy players with the duplicates

```python
vandy_players = cp_salaries_people_df_filtered[(cp_salaries_people_df_filtered['schoolid']=='vandy')]
vandy_players
```

### Dropped duplicates, reset index

```python
all_vandy_players_df = vandy_players.drop_duplicates(subset=["playerid"])
all_vandy_players_df = all_vandy_players_df.reset_index(drop=True)
all_vandy_players_df
```
